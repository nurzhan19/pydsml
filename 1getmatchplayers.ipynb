```python
import re
import requests
import datetime
from bs4 import BeautifulSoup
from python_utils import converters
import json
import pytz
from datetime import datetime

import datetime
from datetime import timedelta, datetime
from dateutil.relativedelta import relativedelta

def get_parsed_page(url):
    # This fixes a blocked by cloudflare error i've encountered
    headers = {
        "referer": "https://www.hltv.org/stats",
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
    }

    return BeautifulSoup(requests.get(url, headers=headers).text, "lxml")
```

```python
def get_matches():
    PAGE = get_parsed_page("http://www.hltv.org/")
    #DIV top-border-hide
    result= []
    team_list=[]

    rez = PAGE.find("div", {"class": "top-border-hide"}).find_all("a", {})
    for item in rez:
        result.append((item.attrs)['href'].split("/")[2])

    return result
```

```python
##########################################################################################################################
##########################################################################################################################

def generateurl(player_id, period, maptype): #player_id, period, maptype
    startDate = datetime.now(pytz.timezone('US/Central'))
    endDate = datetime.now(pytz.timezone('US/Central'))
    #%Y-%m-%d
    if period == "1" and maptype != "all":
        date_after_month = datetime.today()+ relativedelta(months=-1)
        endDate = datetime.today().strftime('%Y-%m-%d')
        startDate = date_after_month.strftime('%Y-%m-%d')
        url = f"/{player_id}/a?startDate={startDate}&endDate={endDate}&maps={maptype}"

    elif period == "3" and maptype != "all":
        date_after_month = datetime.today()+ relativedelta(months=-3)
        endDate = datetime.today().strftime('%Y-%m-%d')
        startDate = date_after_month.strftime('%Y-%m-%d')
        url = f"/{player_id}/a?startDate={startDate}&endDate={endDate}&maps={maptype}"

    elif period == "6" and maptype != "all":
        date_after_month = datetime.today()+ relativedelta(months=-6)
        endDate = datetime.today().strftime('%Y-%m-%d')
        startDate = date_after_month.strftime('%Y-%m-%d')
        url = f"/{player_id}/a?startDate={startDate}&endDate={endDate}&maps={maptype}"

    elif period == "12" and maptype != "all":
        date_after_month = datetime.today()+ relativedelta(months=-12)
        endDate = datetime.today().strftime('%Y-%m-%d')
        startDate = date_after_month.strftime('%Y-%m-%d')
        url = f"/{player_id}/a?startDate={startDate}&endDate={endDate}&maps={maptype}"

    elif period == "1" and maptype == "all":
        date_after_month = datetime.today()+ relativedelta(months=-1)
        endDate = datetime.today().strftime('%Y-%m-%d')
        startDate = date_after_month.strftime('%Y-%m-%d')
        url = f"/{player_id}/a?startDate={startDate}&endDate={endDate}"

    elif period == "3" and maptype == "all":
        date_after_month = datetime.today()+ relativedelta(months=-3)
        endDate = datetime.today().strftime('%Y-%m-%d')
        startDate = date_after_month.strftime('%Y-%m-%d')
        url = f"/{player_id}/a?startDate={startDate}&endDate={endDate}"

    elif period == "6" and maptype == "all":
        date_after_month = datetime.today()+ relativedelta(months=-6)
        endDate = datetime.today().strftime('%Y-%m-%d')
        startDate = date_after_month.strftime('%Y-%m-%d')
        url = f"/{player_id}/a?startDate={startDate}&endDate={endDate}"

    elif period == "12" and maptype == "all":
        date_after_month = datetime.today()+ relativedelta(months=-12)
        endDate = datetime.today().strftime('%Y-%m-%d')
        startDate = date_after_month.strftime('%Y-%m-%d')
        url = f"/{player_id}/a?startDate={startDate}&endDate={endDate}"

    elif period == "all" and maptype != "all":
        date_after_month = datetime.today()+ relativedelta(months=-12)
        endDate = datetime.today().strftime('%Y-%m-%d')
        startDate = date_after_month.strftime('%Y-%m-%d')
        url = f"/{player_id}/a?startDate=all&maps={maptype}"

    else:
        url = f"/{player_id}/a"

    return url

##########################################################################################################################
##########################################################################################################################
```

```python
def get_players(match_id):
    players = []
    players1 = []
    players2 = []
    players_div = {}
    page = get_parsed_page(f"http://www.hltv.org/matches/{match_id}/a")

    players_div = page.find_all("td", {"class": "player player-image"})
    team_list = []
    for item in players_div:
        id_p = item.find({'div': {'class': 'player-compare'}}).attrs
        team = {
            "id": id_p['data-player-id']
        }
        team_list.append(team)
    return team_list
```

```python
def getmapsfrommatch(match_id):
    page = get_parsed_page(f"http://www.hltv.org/matches/{match_id}/a")
    map_div = page.find_all("div", {"class": "mapname"})


    #DIV top-border-hide
    result= []
    team_list=[]
    count=0
    for item in map_div:
        result.append((item).text)
        count=count+1
        if count ==5:
            break


    return result
```
